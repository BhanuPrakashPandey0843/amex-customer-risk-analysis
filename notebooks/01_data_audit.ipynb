## Load Dataset

```python
# Define paths
project_root = Path.cwd().parent
data_path = project_root / "data" / "raw" / "CFPB Complaints Data - Jan25 to Mar26.xlsx"
output_dir = project_root / "data" / "processed"

# Load data
df, metadata = load_data(data_path)

print(f"Dataset loaded: {metadata['rows']} rows, {metadata['columns']} columns")
print(f"Memory usage: {metadata['memory_usage']:.2f} MB")
```
