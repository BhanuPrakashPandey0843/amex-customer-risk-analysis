# Notebook Purpose
Transform raw CFPB complaint data into analysis-ready dataset using enterprise cleaning pipeline.

---

# 1. Setup & Imports

```python
import pandas as pd
import numpy as np
from pathlib import Path
import sys
sys.path.insert(0, str(Path.cwd().parent / 'src'))

from data_cleaning import load_data, clean_dates, standardize_categories, engineer_features, validate_data

print('Imports complete!')
```

---

# 2. Load Raw Data

## Objective
Load raw CFPB dataset and verify dimensions.

## Business Implication
Starting with complete, unaltered data ensures reproducibility.

```python
project_root = Path.cwd().parent
raw_data_path = project_root / 'data' / 'raw' / 'CFPB Complaints Data - Jan25 to Mar26.xlsx'
df_raw = load_data(raw_data_path)

print(f'Raw data loaded: {len(df_raw):,} rows, {len(df_raw.columns)} columns')
```

---

# 3. Text Cleaning & Initial Prep

## Objective
Clean whitespace and normalize text.

## Business Implication
Consistent text formatting ensures accurate comparisons.

```python
df_clean = df_raw.copy()
for col in df_clean.columns:
    if pd.api.types.is_string_dtype(df_clean[col]) or pd.api.types.is_object_dtype(df_clean[col]):
        df_clean[col] = df_clean[col].str.strip()

print('Text cleaning complete')
```

---

# 4. Date Cleaning

## Objective
Parse dates to datetime format for time-based analysis.

## Business Implication
Date-based trends are critical for executive decision-making.

```python
df_clean = clean_dates(df_clean)

print('Date cleaning complete')
print(f"Date range: {df_clean['Date received'].min()} to {df_clean['Date received'].max()}")
```

---

# 5. Categorical Standardization

## Objective
Standardize company names and handle missing values.

## Business Implication
Readable company names improve competitor benchmarking clarity.

```python
df_clean = standardize_categories(df_clean)

print('Standardization complete')
print(f"Unique companies: {df_clean['Company'].nunique()}")
print(f"Companies: {df_clean['Company'].unique().tolist()}")
```

---

# 6. Feature Engineering

## Objective
Create analytical features for deeper insights.

## Business Implication
Engineered features enable trend, seasonality, and performance analysis.

```python
df_clean = engineer_features(df_clean)

print('Feature engineering complete')
print(f"Columns after: {len(df_clean.columns)}")
print(f"New columns added: {[col for col in df_clean.columns if col not in df_raw.columns]}")
```

---

# 7. Data Validation

## Objective
Verify dataset integrity.

## Business Implication
Executive insights depend on validated, trustworthy data.

```python
is_valid, issues = validate_data(df_clean)

if is_valid:
    print('✅ ALL VALIDATIONS PASSED!')
else:
    print(f'❌ VALIDATION FAILED: {issues}')
```

---

# 8. Export Clean Dataset

## Objective
Save analysis-ready dataset.

## Business Implication
Clean, standardized dataset enables reproducible analysis for all phases.

```python
output_path = project_root / 'data' / 'processed' / 'complaints_cleaned.csv'
df_clean.to_csv(output_path, index=False)

print(f'Clean dataset exported to {output_path}')
print(f"Final dimensions: {len(df_clean):,} rows, {len(df_clean.columns)} columns")
```
